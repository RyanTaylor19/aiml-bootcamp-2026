firstly, import all the important libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

create your three sentences

```python
sentences = [
    "In a solitary style and crash the cemetery gates",
    "I have been waiting for this",
    "The weather is terrible"
]
```

create a tokenizer

```python
tokenizer = Tokenizer()
```

fit the tokenizer

```python
tokenizer.fit_on_texts(sentences)
```

see the vocabulary

```python
print(tokenizer.word_index)
```

convert sentences into vectors

```python
sequences = tokenizer.texts_to_sequences(sentences)

print(sequences)
```

pad the sequences

```python
padded_sequences = pad_sequences(sequences, padding='post')

print(padded_sequences)
```

create embedding layer

```python
embedding_layer = tf.keras.layers.Embedding(
    input_dim=len(tokenizer.word_index)+1,
    output_dim=8
)
```

convert vectors into embeddings

```python
embeddings = embedding_layer(padded_sequences)

print(embeddings)
```

assign sentiment labels

```python
labels = np.array([1,1,0])
```

build a sentiment model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(
        input_dim=len(tokenizer.word_index)+1,
        output_dim=8
    ),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

compile the model

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

train the model

```python
model.fit(
    padded_sequences,
    labels,
    epochs=30
)
```

predict sentiment/ predict a new sentence

```python
new_sentence = [
    "the awful things that i have seen"
]

new_sequence = tokenizer.texts_to_sequences(new_sentence)

new_padded = pad_sequences(
    new_sequence,
    maxlen=padded_sequences.shape[1],
    padding='post'
)

prediction = model.predict(new_padded)
```

```python
if prediction[0][0] > 0.5:
    print("Positive")
else:
    print("Negative")
```
